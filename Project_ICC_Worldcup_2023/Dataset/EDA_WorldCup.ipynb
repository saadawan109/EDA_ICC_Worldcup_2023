### Create A Function which Calculate Total Runs, Runs Type , Dismmisals AND Wicket Type of A Player :

```python
def striker_stats(df, striker_name):
   
    striker_df = df[df['striker'] == striker_name]
    
    if striker_df.empty:
        return f"No data available for {striker_name}", pd.DataFrame() 

    total_runs = striker_df['runs_off_bat'].sum()

    sixes = (striker_df['runs_off_bat'] == 6).sum()
    fours = (striker_df['runs_off_bat'] == 4).sum()
    runs_0 = (striker_df['runs_off_bat'] == 0).sum()
    runs_1 = (striker_df['runs_off_bat'] == 1).sum()
    runs_2 = (striker_df['runs_off_bat'] == 2).sum()
    runs_3 = (striker_df['runs_off_bat'] == 3).sum()

    
    dismissals = striker_df[striker_df['wicket_type'].notnull()]
    times_dismissed = dismissals.shape[0]
    dismissal_types = dismissals['wicket_type'].value_counts()

    
    summary_striker = pd.DataFrame({
        'Total Runs': [total_runs],
        'Sixes': [sixes],
        'Fours': [fours],
        'Missed Ball': [runs_0],
        'Runs (1)': [runs_1],
        'Runs (2)': [runs_2],
        'Runs (3)': [runs_3],
        'Times Dismissed': [times_dismissed]
    })

    # Append dismissal types to the summary DataFrame
    for dismissal in dismissal_types.index:
        summary_striker[dismissal] = [dismissal_types[dismissal]]

    return summary_striker

```

### Summary of Batting V Kohli :

```python
striker_name = 'V Kohli'  
summary = striker_stats(df, striker_name)

print(f"Summary Statistics of {striker_name} : ")
print(summary.to_string(index=False))

summary_VKohli = summary.copy()
```

### Create A Function who return Bowlers Wickets, runs, runs type AND wicket, wicket type :

```python
def bowler_statistics(bowler_name, df):

    bowler_data = df[df['bowler'] == bowler_name]
    total_wickets = bowler_data['wicket_type'].count()
    ducks = bowler_data[bowler_data['runs_off_bat'] == 0]
    duck_balls = ducks.shape[0]

    runs_by_bat = {
        '6s': bowler_data[bowler_data['runs_off_bat'] == 6].shape[0],
        '4s': bowler_data[bowler_data['runs_off_bat'] == 4].shape[0],
        '3s': bowler_data[bowler_data['runs_off_bat'] == 3].shape[0],
        '2s': bowler_data[bowler_data['runs_off_bat'] == 2].shape[0],
        '1s': bowler_data[bowler_data['runs_off_bat'] == 1].shape[0],
    }
    
    wicket_types = bowler_data['wicket_type'].value_counts().to_dict()
    
    summary_data = {
        'Bowler': [bowler_name],
        'Total Wickets': [total_wickets],
        'Duck Balls': [duck_balls],
        **runs_by_bat,
    }
    
    for wicket_type in wicket_types:
        summary_data[wicket_type] = [wicket_types[wicket_type]]


    summary_df = pd.DataFrame(summary_data)

    return summary_df
```

```python
bowler_name = 'Shaheen Shah Afridi'  
summary = bowler_statistics(bowler_name, df)

print(f"Summary Statistics of {bowler_name}:")
print(summary.to_string(index=False))

summary_SS_Afridi = summary.copy()
```

```python
bowler_name = 'JJ Bumrah'  
summary = bowler_statistics(bowler_name, df)

print(f"Summary Statistics of {bowler_name}:")
print(summary.to_string(index=False))

summary_JJ_Bumrah = summary.copy()
```

```python
total_wickets_afridi = summary_SS_Afridi['Total Wickets'].values[0]
total_wicket_bumrah = summary_JJ_Bumrah['Total Wickets'].values[0]

labels = ['Afridi', 'Bumrah']
total_wickets = [total_wickets_afridi, total_wicket_bumrah]

plt.figure(figsize=(8, 6))
plt.bar(labels, total_wickets)

plt.title('Total Wickets Comparison: Afridi vs Bumrah', fontsize=16)
plt.xlabel('Bowlers', fontsize=14)
plt.ylabel('Total Wickets', fontsize=14)

for index, value in enumerate(total_wickets):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=12)

plt.show()
```
