### Complete Exploratory Data Analysis :

## About DataSet :

- `Description` :
The Data Set was downloaded from Kaggle, from the following [link](https://www.kaggle.com/datasets/gauthamp10/google-playstore-apps/data)
- `Context` :
While many public datasets (on Kaggle and the like) provide Apple App Store data, there are not many counterpart datasets available for Google Play Store apps anywhere on the web. On digging deeper, I found out that iTunes App Store page deploys a nicely indexed appendix-like structure to allow for simple and easy web scraping. On the other hand, Google Play Store uses sophisticated modern-day techniques (like dynamic page load) using JQuery making scraping more challenging.
- `Content` :
Each app (row) has values for catergory, rating, size, and more.
- `Acknowledgements` :
This information is scraped from the Google Play Store. This app information would not be available without it.
- `Inspiration` :
The Play Store apps data has enormous potential to drive app-making businesses to success. Actionable insights can be drawn for developers to work on and capture the Android market!

### What is the distribution of app ratings across different categories?

```python
avg_rating = df.groupby('Category')['Rating'].mean().reset_index()
avg_rating = avg_rating.sort_values(by='Rating', ascending=False).head(10)
avg_rating
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Rating', y='Category', data=avg_rating, palette='viridis')
plt.title('Average App Ratings by Category')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.grid(axis='x')
plt.show()
```

### Which app categories have the highest number of installs?

```python
# Convert the 'Installs' column to numeric, removing any unwanted characters (e.g., commas, plus signs)
df['Installs'] = df['Installs'].str.replace('+', '').str.replace(',', '').astype(int)
```

```python
install_counts = df.groupby('Category')['Installs'].sum().reset_index()
install_counts = install_counts.sort_values(by='Installs', ascending=False)
```

```python
install_counts = install_counts.head(10)
install_counts
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Installs', y='Category', data=install_counts, palette='viridis')
plt.title('Total Installs by Category')
plt.xlabel('Total Installs')
plt.ylabel('Category')
plt.xticks(rotation=90)
tick_positions = [10_000_000_000, 20_000_000_000, 30_000_000_000, 40_000_000_000, 50_000_000_000, 60_000_000_000, 70_000_000_000]
tick_labels = ["10B", "20B", "30B", "40B", "50B", "60B", "70B"]

plt.xticks(ticks=tick_positions, labels=tick_labels)
plt.grid(axis='x')

plt.show()
```

### What is the relationship between the number of reviews and app ratings?

```python
df['Rating'].head(10)
```

### How does the price of apps influence their ratings?

```python
df.groupby('Price')['Rating'].mean()
```

```python
bins = [0, 1, 2, 5, 10, 20, 50, 100]
labels = ['Free', '0-1$', '1-2$', '2-5$', '5-10$', '10-20$', '20+']
df['Price Range'] = pd.cut(df['Price'], bins=bins, labels=labels)

# Create the box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Price Range', y='Rating', data=df, palette='viridis')
plt.title('App Ratings by Price Range')
plt.xlabel('Price Range (in $)')
plt.ylabel('App Rating')
plt.grid(axis='y')

plt.show()
```

### What are the trends in app releases over time?

```python
df['Released'].head(10)
```

```python
df['Released'] = pd.to_datetime(df['Released'], errors='coerce')  
```

```python
df['Year'] = df['Released'].dt.year
```

```python
release_trend = df.groupby('Year').size().reset_index(name='Number of Releases')
release_trend
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Number of Releases', data=release_trend, marker='o')
plt.title('Trend in App Releases Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Apps Released')
plt.xticks(release_trend['Year'], rotation=45)
plt.grid()

plt.show()
```

### Which developers have the most apps published, and what are their average ratings?

#### Top 10 developrs :

```python
top_developers = df['Developer Id'].value_counts().reset_index()
top_developers = top_developers.head(10)
top_developers
```

```python
plt.Figure(figsize=(12,6))
sns.lineplot(y='count',x='Developer Id', data=top_developers)
plt.title('Top 10 Developers by Number of Apps Published')
plt.xlabel('Number of Apps')
plt.ylabel('Developer Id')
plt.grid(axis='x')
plt.xticks(rotation=90)

plt.show()
```

### Their Avg Ratings :

```python
rating_top_developers = df[df['Developer Id'].isin(top_developers['Developer Id'])]
```

```python
average_rating = rating_top_developers.groupby('Developer Id')['Rating'].mean().reset_index()
average_rating.head()
```

```python
top_developers_info = pd.merge(top_developers, average_rating, on='Developer Id')
top_developers_info.columns = ['Developer Id', 'Total Installs', 'Average Rating']
print(top_developers_info)
```

```python
plt.subplot(1, 2, 2)
sns.lineplot(y='Average Rating', x='Developer Id', data=top_developers_info, marker='o')
plt.title('Average Ratings of Top 10 Developers')
plt.xlabel('Average Rating')
plt.ylabel('Developer Id')
plt.grid(axis='x')
plt.xticks(rotation=90)

# Adjust layout
plt.tight_layout()
plt.show()
```
